## Few-Shot Prompting (Smart Prompting)

```python
from langchain_core.prompts import FewShotPromptTemplate

examples = [
    {"input": "I love this product", "output": "Positive"},
    {"input": "This is bad", "output": "Negative"},
]

example_prompt = PromptTemplate(
    input_variables=["input", "output"],
    template="Text: {input}\nSentiment: {output}"
)

few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix="Classify sentiment:\n",
    suffix="Text: {input}\nSentiment:",
    input_variables=["input"]
)

chain = few_shot_prompt | llm | StrOutputParser()

input = "This is amazing"
print("Text:",input,"\nSentiment:",chain.invoke(input))
```

## Embeddings (Gemini)
This section shows how text is converted into vectors and used for similarity search.

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001"
)

vec1 = embeddings.embed_query("cat")
vec2 = embeddings.embed_query("dog")

print(vec1[:5])
```

## Similarity Check

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity([vec1], [vec2])
print("Similarity:", similarity)
```

## Vector Store (FAISS)

```python
from langchain_community.vectorstores import FAISS

texts = ["LangChain is powerful", "AI is transforming the world"]

vectorstore = FAISS.from_texts(texts, embeddings)

results = vectorstore.similarity_search("What is LangChain?")
print(results[0].page_content)
```

## Basic LLM call

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Step 1: Model
model = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

# Step 2: Prompt
prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in simple terms"
)

# Step 3: LCEL Chain
chain = prompt | model | StrOutputParser()

# Step 4: Invoke
response = chain.invoke({"topic": "NLP"})
print(response)
```

## Agent with Tool

```python
from langchain.agents import create_agent
from langchain_core.tools import tool

@tool
def calculate_bmi(weight_kg: float, height_m: float) -> str:
    """Calculate BMI given weight and height."""
    bmi = weight_kg / (height_m ** 2)
    return f"BMI is {bmi:.2f}"

@tool
def celsius_to_fahrenheit(celsius: float) -> str:
    """Convert Celsius to Fahrenheit."""
    return f"{(celsius * 9/5) + 32:.1f} °F"

tools = [calculate_bmi, celsius_to_fahrenheit]

# Create agent (NEW API)
agent = create_agent(
    model=model,
    tools=tools
)

# Run agent
result = agent.invoke({
    "messages": [
        ("human", "Calculate BMI for 80kg, 1.80m and convert 34C to Fahrenheit")
    ]
})

print(result["messages"][-1].content)
```

## Memory

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage

# Memory storage
chat_history = []

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])

chain = prompt | model

def chat(user_input):
    chat_history.append(HumanMessage(content=user_input))

    response = chain.invoke({
        "input": user_input,
        "chat_history": chat_history
    })

    chat_history.append(AIMessage(content=response.content))
    return response.content

# Conversation
print("Q1.","Hi, my name is Meghna. I am doing my internship at Innomatics\n",chat("Hi, my name is Meghna. I am doing my internship at Innomatics"),"\n")
print("-"*60)
print("\nQ2.","Explain API\n",chat("Explain API"),"\n")
print("-"*60)
print("\nQ3.","What is my name and where am I doing my internship?\n",chat("What is my name and where ai I doing my internship?"))
```

```python
print("\n--- Memory Buffer ---")
for msg in chat_history:
    role = "Human" if isinstance(msg, HumanMessage) else "AI"
    print(f"[{role}]: {msg.content[:120]}")
```

## **RAG Pipeline**

### Step 1: Documents + Split

```python
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter

docs = [
    Document(page_content="LangChain is an open-source framework created by Harrison Chase in 2022."),
    Document(page_content="FAISS is used for fast similarity search in vector databases."),
    Document(page_content="RAG retrieves documents and injects them into prompts to reduce hallucination.")
]

splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=20)
chunks = splitter.split_documents(docs)
```

### Step 2: Embeddings + Vector Store

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS

embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")

vector_store = FAISS.from_documents(chunks, embeddings)
retriever = vector_store.as_retriever(search_kwargs={"k": 2})
```

### Step 3: RAG Chain

```python
from langchain_core.runnables import RunnablePassthrough

prompt = ChatPromptTemplate.from_template("""
Answer based only on context:

{context}

Question: {question}
""")

def format_docs(docs):
    return "\n".join(doc.page_content for doc in docs)

rag_chain = (
    {
        "context": retriever | format_docs,
        "question": RunnablePassthrough()
    }
    | prompt | model | StrOutputParser()
)
```

### Step 4: Query

```python
print("Q1. Who created LangChain?\n",rag_chain.invoke("Who created LangChain?"),"\n")
print("Q2. What is FAISS?\n",rag_chain.invoke("What is FAISS?"),"\n")
print("Q3. What is RAG?\n",rag_chain.invoke("What is RAG?"))
```

# **5. Real-World Use Cases**

## USE CASE 1: AI Customer Support Chatbot
**Problem**

Users need quick answers + context-aware responses.

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage

chat_history = []

support_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a professional customer support assistant for an e-commerce company."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])

support_chain = support_prompt | model

def customer_support(query):
    chat_history.append(HumanMessage(content=query))

    response = support_chain.invoke({
        "input": query,
        "chat_history": chat_history
    })

    chat_history.append(AIMessage(content=response.content))
    return response.content

print("-" * 60)
print("SuperStore — Customer Support Chat")
print("-" * 60)

s1 = "My order is delayed, what should I do?"
s2 = "My order ID is 12345"
print("Q1.",s1,"\n",customer_support(s1))
print('-'*40)
print("\nQ2.",s2,"\n",customer_support(s2))
```

## USE CASE 2: Document Q&A System (RAG)
**Problem**

Users want answers from documents (not general AI knowledge)

```python
from langchain_core.documents import Document

docs = [
    Document(page_content="LangChain was created by Harrison Chase in 2022."),
    Document(page_content="RAG improves accuracy by retrieving relevant documents."),
    Document(page_content="FAISS is used for fast similarity search.")
]

# Split
splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=20)
chunks = splitter.split_documents(docs)

# Vector Store
vector_store = FAISS.from_documents(chunks, embeddings)
retriever = vector_store.as_retriever()

# RAG Prompt
prompt = ChatPromptTemplate.from_template("""
Answer ONLY from context:

{context}

Question: {question}
""")

def format_docs(docs):
    return "\n".join(doc.page_content for doc in docs)

rag_chain = (
    {
        "context": retriever | format_docs,
        "question": RunnablePassthrough()
    }
    | prompt | model | StrOutputParser()
)

print("Q1. Who created LangChain?\n",rag_chain.invoke("Who created LangChain?"))
```

## USE CASE 3: AI Research Assistant
**Problem**

```python
from langchain.agents import create_agent
from langchain_core.tools import tool

@tool
def summarize_text(text: str) -> str:
    """Summarize the given text."""
    return f"Summary: {text[:100]}..."

tools = [summarize_text]

research_agent = create_agent(
    model=model,
    tools=tools
)

result = research_agent.invoke({
    "messages": [
        ("human", "Explain recent trends in AI and summarize them")
    ]
})

print(result["messages"][-1].content)
```
